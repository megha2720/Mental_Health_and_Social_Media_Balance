**Converting Categorical Data to Numberical Data**

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
for column in df.columns:
    df[column] = LE.fit_transform(df[column])
```

```python
df.info()
```

## Modeling

```python
final_res = []
```

```python
x = df.drop(['User_ID','Happiness_Index(1-10)'], axis = 1)
y = df['Happiness_Index(1-10)']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

**Linear Regression**

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x_train,y_train)
```

```python
y_pred_LR = LR.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
print('Linear Regression:')
print('R2 Score :',r2_score(y_test,y_pred_LR))
print('Mean Absolute Error :',mean_absolute_error(y_test,y_pred_LR))
r2_score_LR = r2_score(y_test,y_pred_LR)
final_res.append(r2_score_LR)
```

**Decision Tree** 

```python
from sklearn.tree import DecisionTreeRegressor
DT = DecisionTreeRegressor()
DT.fit(x_train,y_train)
```

```python
y_pred_DT = DT.predict(x_test)
```

```python
print('Decision Tree :')
print('R2 Score :',r2_score(y_test,y_pred_DT))
print('Mean Absolute Error :',mean_absolute_error(y_test,y_pred_DT))
r2_score_DT = r2_score(y_test,y_pred_DT)
final_res.append(r2_score_DT)
```

**Random Forest Regressor**

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor()
RF.fit(x_train,y_train)
```

```python
y_pred_RF = RF.predict(x_test)
```

```python
print('Random Forest Regressor :')
print('R2 Score :',r2_score(y_test,y_pred_RF))
print('Mean Absolute Error :',mean_absolute_error(y_test,y_pred_RF))
r2_score_RF = r2_score(y_test,y_pred_RF)
final_res.append(r2_score_RF)
```

**XGBoost**

```python
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(x_train,y_train)
```

```python
y_pred_xgb = xgb.predict(x_test)
```

```python
print('XGBoost :')
print('R2 Score :', r2_score(y_test,y_pred_xgb))
print('Mean Absolute Error :', mean_absolute_error(y_test,y_pred_xgb))
r2_score_xgb = r2_score(y_test,y_pred_xgb)
final_res.append(r2_score_xgb)
```

**K-Nearest Neighbors**

```python
from sklearn.neighbors import KNeighborsRegressor
KNN = KNeighborsRegressor()
KNN.fit(x_train,y_train)
```

```python
y_pred_KNN = KNN.predict(x_test)
```

```python
print('K-Nearest Neighbors :')
print('R2 Score :', r2_score(y_test,y_pred_KNN))
print(' Mean Absolute Error :', mean_absolute_error(y_test,y_pred_KNN))
r2_score_KNN = r2_score(y_test,y_pred_KNN)
final_res.append(r2_score_KNN)
```

**Gradient Boosting**

```python
from sklearn.ensemble import GradientBoostingRegressor
GBC = GradientBoostingRegressor()
GBC.fit(x_train,y_train)
```

```python
y_pred_GBC = GBC.predict(x_test)
```

```python
print('Gradient Boosting ')
print('R2 Score ',r2_score(y_test,y_pred_GBC))
acc_GBC = r2_score(y_test,y_pred_GBC)
final_res.append(acc_GBC)
```

**Support Vector Machine**

```python
from sklearn.svm import SVC
SV = SVC()
SV.fit(x_train,y_train)
```

```python
y_pred_SV = SV.predict(x_test)
```

```python
print('Support Vector Machine ')
print('R2 Score ',r2_score(y_test,y_pred_SV))
acc_SV = r2_score(y_test,y_pred_SV)
final_res.append(acc_SV)
```

## Comparing Algorithm

```python
plt.xticks(rotation = 90)
sns.barplot(x = ['Linear Regression','Decision Tree', 'Random Forest','XGBoost','K-Nearest Neighbors','Gradient Boosting ','Support Vector Machine '], y = [r2_score(y_test, y_pred_LR), r2_score(y_test,y_pred_DT), r2_score(y_test, y_pred_RF), r2_score(y_test, y_pred_KNN), r2_score(y_test, y_pred_xgb), r2_score(y_test, y_pred_GBC),r2_score(y_test,y_pred_SV)]).set_title('R2 Score')
```

```python
final=np.array(final_res)
result=final.reshape(-1,1)
columns=['R2 Score']
index=['Linear Regression','Decision Tree', 'Random Forest','XGBoost','K-Nearest Neighbors','Gradient Boosting ','Support Vector Machine ']
final_result=pd.DataFrame(result,index=index,columns=columns)
final_result.sort_values(by='R2 Score', ascending=False)
```

## Hyperparameter optimization

#### Tuning Best Performing Models

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
```

**Random Forest**

```python
grid_rf = RandomForestRegressor(random_state=42)

param_rf = {
    'n_estimators': [100, 200, 300, 500],     # Number of trees
    'max_depth': [None, 10, 20, 30, 40],      # Maximum depth of each tree
    'min_samples_split': [2, 5, 10],          # Minimum samples to split a node
    'min_samples_leaf': [1, 2, 4],            # Minimum samples at a leaf node
    'max_features': ['auto', 'sqrt', 'log2'], # Number of features to consider per split
    'bootstrap': [True, False]                # Whether bootstrap samples are used
}

rf_search = RandomizedSearchCV(
    estimator=grid_rf,
    param_distributions=param_rf,
    n_iter=10,       # number of random combinations to try
    cv=3,            # 3-fold cross-validation
    scoring='r2',    # optimize for R
    n_jobs=-1,       # use all cores
    random_state=42
)

# 4 Fit the model
rf_search.fit(x_train, y_train)
```

```python
y_pred_rf = rf_search.predict(x_test) 
```

```python
print('R2 Score: ',r2_score(y_test,y_pred_rf))
print('Mean Absolute Score: ',mean_absolute_error(y_test,y_pred_rf))
r2_score_model_rf = r2_score(y_test,y_pred_rf)
mae_model_rf = mean_absolute_error(y_test,y_pred_rf)
```

**Gradient Boosting**

```python
gbr = GradientBoostingRegressor(random_state=42)
```

```python
param_grid = {
    'n_estimators': [100, 200, 300, 500],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'subsample': [0.8, 1.0]
}

grid_gbr = GridSearchCV(
    estimator=gbr,
    param_grid=param_grid,
    cv=3,                  # 5-fold cross-validation
    n_jobs=-1,             # use all CPU cores
    scoring='r2',          # evaluation metric
    verbose=2
)

grid_gbr.fit(x_train, y_train)
```

```python
y_pred_gbr = grid_gbr.predict(x_test)
```

```python
print("R2 Score:", r2_score(y_test,y_pred_gbr))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred_gbr))
r2_score_model_gbr = r2_score(y_test,y_pred_gbr)
mae_model_gbr = mean_absolute_error(y_test,y_pred_gbr)
```

```python
models_score = pd.DataFrame({'R2 Score':[r2_score_model_gbr,r2_score_model_rf],'Mean Absolute Error':[mae_model_gbr,mae_model_rf]}, index=['Gradient Boosting','Random Forest'])
models_score
```
